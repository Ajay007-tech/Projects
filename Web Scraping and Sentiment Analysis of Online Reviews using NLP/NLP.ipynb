```python
nltk.download('punkt')
nltk.download('cmudict')

# Function to load stopwords from multiple files
def load_stopwords(stopwords_files):
    stopwords_list = set()
    for file in stopwords_files:
        with open(file, 'r', encoding='latin-1') as f:
            stopwords_list.update(f.read().splitlines())
    return stopwords_list

# Function to create positive and negative word dictionaries from Master Dictionary files
def create_word_dictionary(positive_file, negative_file, stopwords_list):
    positive_dict = {}
    negative_dict = {}

    # Load positive words
    with open(positive_file, 'r') as f:
        for line in f:
            word = line.strip()
            if word not in stopwords_list:
                positive_dict[word] = 1  # Assigning a score of 1 for positive words

    # Load negative words
    with open(negative_file, 'r',encoding='latin-1') as f:
        for line in f:
            word = line.strip()
            if word not in stopwords_list:
                negative_dict[word] = -1  # Assigning a score of -1 for negative words

    return positive_dict, negative_dict

# Function to clean text by removing punctuation, converting to lowercase, and removing stopwords
def clean_text(text, stopwords_list):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Remove punctuation marks
    table = str.maketrans('', '', string.punctuation)
    words = [word.translate(table) for word in words]

    # Convert words to lowercase
    words = [word.lower() for word in words]

    # Remove stopwords
    words = [word for word in words if word not in stopwords_list]

    # Join the words back into a single string
    cleaned_text = ' '.join(words)

    return cleaned_text

# Function to count syllables in a word
def count_syllables(word):
    # Count syllables based on number of vowels in word
    count = 0
    vowels = "aeiouy"
    word = word.lower()
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith("e"):
        count -= 1
    if word.endswith("le") and len(word) > 2 and word[-3] not in vowels:
        count += 1
    if count == 0:
        count += 1
    return count

# Function to count personal pronouns
def count_personal_pronouns(text):
    personal_pronouns = ['i', 'we', 'my', 'ours', 'us']
    pronoun_count = 0
    # Match personal pronouns in text
    for pronoun in personal_pronouns:
        pronoun_count += len(re.findall(r'\b' + pronoun + r'\b', text.lower()))
    return pronoun_count

# Function to calculate average word length
def average_word_length(text):
    # Remove punctuation marks
    text = re.sub(r'[^\w\s]', '', text)
    words = word_tokenize(text)
    total_chars = sum(len(word) for word in words)
    total_words = len(words)
    if total_words != 0:
        return total_chars / total_words
    else:
        return 0

# Path to the Master Dictionary files
positive_file = "positive-words.txt"
negative_file = "negative-words.txt"

# List of stopwords files
stopwords_files = ['StopWords_Names.txt', 'StopWords_Geographic.txt', "StopWords_Auditor.txt",
                   'StopWords_GenericLong.txt', 'StopWords_Generic.txt', 'StopWords_DatesandNumbers.txt',
                   'StopWords_Currencies.txt']

# Load stopwords from all files
stopwords_list = load_stopwords(stopwords_files)

# Create positive and negative word dictionaries
positive_dict, negative_dict = create_word_dictionary(positive_file, negative_file, stopwords_list)

# Process each file
output_data = []

for i in df["URL_ID"]:
    filename = i + ".txt"
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            text = f.read()
        cleaned_text = clean_text(text, stopwords_list)

        # Initialize positive and negative scores
        positive_score = 0
        negative_score = 0

        # Calculate scores for positive and negative words
        for word in cleaned_text.split():
            positive_score += positive_dict.get(word, 0)
            negative_score += negative_dict.get(word, 0)

        # Count syllables per word
        syllable_count = sum(count_syllables(word) for word in cleaned_text.split())

        # Count personal pronouns
        personal_pronoun_count = count_personal_pronouns(cleaned_text)

        # Calculate average word length
        avg_word_len = average_word_length(cleaned_text)

        # Calculate polarity score
        polarity_score = (positive_score - (negative_score*-1))/ ((positive_score +(negative_score)*-1) + 0.000001)

        # Calculate subjectivity score
        subjectivity_score = (positive_score  + negative_score)/ ((len(cleaned_text)) + 0.000001)

        # Count total number of sentences
        total_sentences = len(sent_tokenize(text))

        # Count total number of words
        total_words = len(cleaned_text.split())

        # Calculate average sentence length
        avg_sentence_len = total_words / total_sentences

        # Calculate percentage of complex words
        percentage_complex_words = syllable_count / total_words

        # Calculate Fog Index
        fog_index = 0.4 * (avg_sentence_len + percentage_complex_words)

        # Append data to output list
        output_data.append({
            "URL_ID": i,
            "POSITIVE SCORE": positive_score,
            "NEGATIVE SCORE": negative_score,
            "POLARITY SCORE": polarity_score,
            "SUBJECTIVITY SCORE": subjectivity_score,
            "AVG SENTENCE LENGTH": avg_sentence_len,
            "PERCENTAGE OF COMPLEX WORDS": percentage_complex_words,
            "FOG INDEX": fog_index,
            "AVG NUMBER OF WORDS PER SENTENCE": total_words / total_sentences,
            "COMPLEX WORD COUNT": syllable_count,
            "WORD COUNT": total_words,
            "SYLLABLE PER WORD": syllable_count / total_words,
            "PERSONAL PRONOUNS": personal_pronoun_count,
            "AVG WORD LENGTH": avg_word_len
        })

# Convert output data to dataframe
output_df = pd.DataFrame(output_data)

# Display the dataframe
print(output_df)
```
